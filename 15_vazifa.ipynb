# Dataset haqida

Ushbu ma'lumotlar to'plami turli kollej va universitetlar, jumladan, ularning joylashuvi, talabalar demografiyasi, moliyaviy jihatlari va institutsional xususiyatlari haqida to'liq ma'lumot beradi. Ushbu ma'lumotlar turli tahlillar uchun ishlatilishi mumkin, masalan, institutlarni qabul stavkalari, o'qish to'lovlari, talabalar qarzi va geografik taqsimoti asosida taqqoslash.

## Ustunlar haqida ma'lumot:

1) INSTNM: Institution Name - Kollej yoki universitet nomi.

2) OPEID: Office of Postsecondary Education Identification - Ta'lim yo'nalishi ID raqami

3) REGION: Geographic Region - Kollej yoki universitet joylashgan hudud.

4) SAT_AVG_ALL: Average SAT Scores - Institutga qabul qilingan barcha talabalarning o'rtacha SAT ballari

5) PCTPELL: Percentage of Students Receiving Pell Grants - Odatda kam ta'minlangan talabalarga beriladigan Pell Grantlarini oladigan talabalari ulushi.

6) PCTFLOAN: Percentage of Students Receiving Federal Loans - Federal kreditlarni olgan talabalar ulushi

7) ADM_RATE_ALL: Admission Rate - Qabul qilingan abituriyentlarning foizini ko'rsatadigan muassasaning umumiy qabul darajasi.

8) UG: Undergraduate Enrollment - Institutga qabul qilingan talabalarning umumiy soni.

9) AVGFACSAL: Average Faculty Salary - O'qituvchilarning o'rtacha ish haqi

10) COMPL_RPY_5YR_RT: 5-Year Loan Repayment Rate - Besh yil ichida o'z talaba kreditlarini muvaffaqiyatli to'lagan talabalar ulushi

11) DEBT_MDN: Median Debt - Institutda o'qigan talabalar tomonidan to'plangan o'rtacha qarz.

12) MEDIAN_HH_INC: Median Household Income - Uy xo'jaliklarining o'rtacha daromadi

13) ICLEVEL: Institution Level - Institut darajasi

14) HIGHDEG: Highest Degree Offered - Taklif etilgan eng yuqori daraja yoki Muassasa tomonidan taqdim etiladigan darajalar turlari 

15) CONTROL: Control of Institution - Muassasa davlat, xususiy notijorat yoki xususiy notijorat ekanligini ko'rsatadi.

16) WOMENONLY: Women Only Institution - Faqat ayollar instituti

17) MENONLY: Men Only Institution - Faqat erkaklar instituti

18) LOCALE: Locale Description - Tashkilotning geografik joylashuvini ifodalovchi kod (masalan, shahar, shahar atrofi, shaharcha, qishloq).

19) Tuition: Tuition Fees - Ta'lim muassasasi tomonidan olinadigan to'lovlar.

20) Degree_Type: Type of Degrees Offered - Taklif etilgan eng yuqori daraja yoki Muassasa tomonidan taqdim etiladigan darajalar turlari = HIGHDEG

21) Ownership: Ownership Type - Tashkilotning mulk turi (masalan, davlat, xususiy). = CONTROL

22) Regions: Geographic Regions = REGION

23) Locales: Locale Categories = LOCALE

24) Locale_Short: Short Locale Description - Geografik joyning qisqartirilgan tavsifi.

### "DEBT_MDN" ustun ustida ishlash

```python
univer["DEBT_MDN_NUMERIC"]=pd.to_numeric(univer["DEBT_MDN"], errors="coerce")
DEBT_MDN_NUMERIC = univer["DEBT_MDN_NUMERIC"]
```

```python
univer["DEBT_MDN_NUMERIC"].hist()
```

```python
univer.groupby(["REGION","Ownership","ICLEVEL"])["DEBT_MDN_NUMERIC"].mean().count()
```

```python
medians_DEBT = univer.groupby(["REGION","Ownership","ICLEVEL"])["DEBT_MDN_NUMERIC"].transform("median")
medians_DEBT
univer["DEBT_MDN_NUMERIC"] = univer["DEBT_MDN_NUMERIC"].fillna(medians_DEBT)
```

```python
univer["DEBT_MDN_NUMERIC"].describe() - pd.to_numeric(univer["DEBT_MDN"], errors="coerce").describe()
```

```python
fig, ax = plt.subplots(figsize=(10,6))
univer["DEBT_MDN_NUMERIC"].plot(kind="hist",ax=ax,alpha=0.5,color="blue",bins=10, label="Fillna_median")
DEBT_MDN_NUMERIC.plot(kind="hist",ax=ax,alpha=0.5,color="green",bins=10, label="Numeric")
ax.set_xlabel("Value")
ax.legend()
plt.grid()
plt.title("Institutda o'qigan talabalar tomonidan to'plangan o'rtacha qarz. Oldingi va Median qilingani")
```

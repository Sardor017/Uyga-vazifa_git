### Index bo'yicha tartiblashni(sort) ko'ramiz

```python
df_BMI.sort_index()
```

```python
df_Age_Gender_Ethnicity_EducationLevel_sorted = df.set_index(["Age","Gender", "Ethnicity", "EducationLevel"])
df_Age_Gender_Ethnicity_EducationLevel_sorted
```

```python
df_Age_Gender_Ethnicity_EducationLevel_sorted.sort_index().head(20)
```

'Age	-> Gender -> Ethnicity	-> EducationLevel'	ketma-ketligi bo'yicha index bo'yicha sortlash

```python
df_AGEE = df_Age_Gender_Ethnicity_EducationLevel_sorted
df_AGEE.sort_index(level = ["Gender", "Age", "Ethnicity"], ascending = [True, False, False]).head(20)
```

```python
df_AGEE.sort_index().loc[7:10]
```

"Age" bo'yicha sort qilib 7 dan 10 yoshgacha bo'lganlarni ko'rish imkoniyatiga ega bo'ldik. Sort qilmagan paytimizda xatolik bo'ldi.

```python
df_AGEE.sort_index().loc[(7,2):(10,1)]
```

"Age" va "Gender" bo'yicha sort qilib 7 dan(ayollar) 10 (erkaklar) yoshgacha bo'lganlarni ko'rish imkoniyatiga ega bo'ldik. Sort qilmagan paytimizda xatolik bo'ldi.

```python
df_AGEE.sort_index().loc[:, "PatientID":"Smoking"]
```

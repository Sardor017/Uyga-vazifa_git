### Funksiya yaratish:

```python
def Nan_to_median(data, column, column_median):
    print("Mediana, O'rtacha, Modani bilib olamiz: ")
    print("O'rtacha: ", data[column].mean())
    print("Mediana: ", data[column].median())
    print("Moda: ", data[column].mode())
    print("Mediana ishlatish maqbul bo'ladi deb o'ylayman")
    sns.displot(data[column], kind="kde", rug=True)
    plt.show()
    print("Gipoteza mizni tekshirib olamiz")
    print(data[data[column].isna()].head())
    print(data[column].value_counts())
    print("'0' juda ham ko'p bo'lganligi uchun 'median' ni ishlatamiz, sababi u '0' ga yaqinroq.")
    print("Yangi ustun hosil qilamiz")
    data[column_median] = data[column].fillna(data[column].median())
    print(data[data[column_median].isna()])
    print("'fillna' metodini ishlatib, medianani qo'yish orqali yangi ustun yaratdik. Uni tekshirib ko'rdik unda 'NaN' boshqa qolmadi.")
    print("Qanday o'zgarishlar bo'ldi tekshirib ko'ramiz (Hist, Std)")
    fig, axes = plt.subplots(2, 1, figsize=(10,10))
    sns.set_context("paper")
    sns.histplot(data[column_median], kde=True, ax=axes[0])
    axes[0].set_title("Medianaga almashtirilgan")
    sns.histplot(data[column].dropna(), kde=True, ax=axes[1])
    axes[1].set_title("NaN lari bor")
    plt.tight_layout()
    plt.show()
    print(data[column].std())
    print(data[column_median].std())
    print("Ko'rib turibmizki grafikda ham, 'Std' da ham o'zgarish deyarli yo'q.")
    print("Boxplot orqali tekshirish")
    sns.boxplot(y=data[column].dropna())
    plt.show()
    sns.boxplot(y=data[column_median])
    plt.show()
    print("Boxplotni ishlatib ham o'zgarish ko'rmadim")
```

```python
def Nan_to_mode(data, column, column_mode):
    print("Modani bilib olamiz: ")
    print("Moda: ", data[column].mode())
    print("Moda ishlatish maqbul bo'ladi deb o'ylayman")
    fig, ax = plt.subplots(figsize=(10,10))
    sns.histplot(data[column], kde=True)
    plt.tight_layout()
    plt.show()
    print("Gipoteza mizni tekshirib olamiz")
    print(data[data[column].isna()].head())
    print(data[column].value_counts())
    print("Yangi ustun hosil qilamiz")
    data[column_mode] = data[column].fillna(data[column].mode()[0])
    print("'fillna' metodini ishlatib, modani qo'yish orqali yangi ustun yaratdik. Uni tekshirib ko'rdik unda 'NaN' boshqa qolmadi.")
    print("Qanday o'zgarishlar bo'ldi tekshirib ko'ramiz (Hist, Std)")
    fig, axes = plt.subplots(2, 1, figsize=(10,10))
    sns.set_context("paper")
    sns.histplot(data[column_mode], kde=True, ax=axes[0])
    axes[0].set_title("Modaga almashtirilgan")
    sns.histplot(data[column].dropna(), kde=True, ax=axes[1])
    axes[1].set_title("NaN lari bor")
    plt.tight_layout()
    plt.show()
    print("Boxplot orqali tekshirish")
    sns.boxplot(y=data[column].dropna())
    plt.tight_layout()
    plt.show()
    sns.boxplot(y=data[column_mode])
    plt.tight_layout()
    plt.show()
    print("Boxplotni ishlatib ham o'zgarish ko'rmadim")
```

```python
hours = planes['Duration'].str.extract(r'(?P<hours>\d+)h')['hours'].fillna(0).astype(int)
minutes = planes['Duration'].str.extract(r'(?P<minutes>\d+)m')['minutes'].fillna(0).astype(int)
# Combine into HH:MM format
planes['Duration_HHMM'] = (hours * 60 + minutes).apply(lambda x: f"{x // 60:02d}:{x % 60:02d}")
planes.head(15)
```

```python
planes[["Date_of_Journey", "Dep_Time",	"Arrival_Time",	"Duration_HHMM"]]
```

```python
planes[["Date_of_Journey", "Dep_Time",	"Arrival_Time",	"Duration_HHMM"]].isna().sum()
```

```python
planes['Date_of_Journey'] = pd.to_datetime(planes['Date_of_Journey'], format='%d/%m/%Y')

# Convert Dep_Time to datetime with Date_of_Journey
planes['Dep_Time'] = pd.to_datetime(planes['Date_of_Journey'].astype(str) + ' ' + planes['Dep_Time'].fillna('00:00'), format='%Y-%m-%d %H:%M', errors='coerce')

# Helper function to parse Arrival_Time
def parse_arrival_time(row):
    try:
        return pd.to_datetime(row['Date_of_Journey'].strftime('%Y-%m-%d') + ' ' + row['Arrival_Time'], format='%Y-%m-%d %H:%M')
    except:
        try:
            return pd.to_datetime(row['Date_of_Journey'].strftime('%Y-%m-%d') + ' ' + row['Arrival_Time'], format='%Y-%m-%d %H:%M')
        except:
            try:
                day_of_journey = row['Date_of_Journey'].day
                return pd.to_datetime(row['Arrival_Time'], format=f'%H:%M %d %b').replace(year=row['Date_of_Journey'].year, month=row['Date_of_Journey'].month, day=day_of_journey)
            except:
                return pd.NaT

planes['Arrival_Time'] = planes.apply(parse_arrival_time, axis=1)

# Convert Duration_HHMM to timedelta
planes['Duration_HHMM'] = pd.to_timedelta(planes['Duration_HHMM'] + ':00')

# Fill in missing Dep_Time
for idx, row in planes[planes['Dep_Time'].isna()].iterrows():
    if pd.notna(row['Arrival_Time']) and pd.notna(row['Duration_HHMM']):
        planes.at[idx, 'Dep_Time'] = row['Arrival_Time'] - row['Duration_HHMM']

# Fill in missing Arrival_Time
for idx, row in planes[planes['Arrival_Time'].isna()].iterrows():
    if pd.notna(row['Dep_Time']) and pd.notna(row['Duration_HHMM']):
        planes.at[idx, 'Arrival_Time'] = row['Dep_Time'] + row['Duration_HHMM']

# Fill in missing Duration_HHMM
for idx, row in planes[planes['Duration_HHMM'].isna()].iterrows():
    if pd.notna(row['Dep_Time']) and pd.notna(row['Arrival_Time']):
        planes.at[idx, 'Duration_HHMM'] = row['Arrival_Time'] - row['Dep_Time']
planes.head(15)
```

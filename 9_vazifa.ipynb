## Point plot

```python
custom_palette = ["red","green","orange","blue","yellow","purple"]
sns.set_palette(custom_palette)
category_names = ecommerce["Category"].unique()
fig, axes = plt.subplots(8, 3, figsize=(20,60), sharey=True)
for ax, category_name  in zip(axes.flatten(), category_names):
    product_name = ecommerce[ecommerce["Category"].isin([category_name])]["ProductName"]
    prices = ecommerce[ecommerce["Category"].isin([category_name])]["Price"]
    catplot = sns.pointplot(x=product_name, y=prices, ax=ax)
    ax.set_title(category_name)
    ax.set_xlabel("Mahsulot nomi")
    ax.set_ylabel("Narxi")
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.tight_layout()
plt.show()
```

## Catplot(Bar)

```python
sns.set_context("notebook")
sns.catplot(x="Category", y="Price", data=ecommerce, kind="box")
plt.xticks(rotation=90)
plt.show()
```

```python
sns.set_context("notebook")
sns.catplot(x="Category", y="Rating", data=ecommerce, kind="box")
plt.xticks(rotation=90)
plt.show()
```

### Hammasini birlashtirish

```python
fig, axes = plt.subplots(2, 2, figsize=(10,10))
sns.set_style("whitegrid")
sns.set_context("paper")
sns.stripplot(data=sales, y="Product Category", x="Unit Price", jitter=True, hue="Region", size=7, ax=axes[0, 0])
axes[0,0].set_title("Mahsulot turi va uning narxi solishtiruvi Davlatlar bo'yicha")
sns.stripplot(data=sales, y="Region", x="Total Revenue", hue="Payment Method", size=6, jitter=True, ax=axes[0,1])
axes[0,1].set_title("Davlatlarda mahsulotlarning umumiy narxi to'lov turi bo'yicha")
sns.stripplot(data=sales, y="Payment Method", x="Total Revenue", hue="Product Category", jitter=True, size=5, ax=axes[1,0])
axes[1,0].set_title("To'lov turi va mahsulotning umumiy narxi mahsulot turi bo'yicha")
plt.tight_layout()
plt.show()
```

Eng qimmat va eng ko'p sotilayotgan narsa electron tovarlarligi yaqqol ko'rinib turibti, undan keyin uy va sport buyumlari.
Tovarlar davlatlar orasida teng taqsimlanmaganidan yoki shu tovarlarni boshqa davlatlarda ham sotilishi ko'rinmayoyganidan biz datasetimiz to'liq emasligini bilib olishimiz mumkin. Amerikada asosan Credit Card dan, Yevropada - PayPal dan, Asia da - Credit Card hamda Debit Card lardan foydalanilishi bilib oldik.

## Swarm plot

```python
fig, axes = plt.subplots(3, 1, figsize=(8, 15))
sns.swarmplot(data=sales, y="Product Category", x="Unit Price", hue="Region", size=7, ax=axes[0])
axes[0].set_title("Mahsulot turi va uning narxi solishtiruvi Davlatlar bo'yicha")
sns.swarmplot(data=sales, y="Region", x="Total Revenue", hue="Payment Method", size=6, ax=axes[1])
axes[1].set_title("Davlatlarda mahsulotlarning umumiy narxi to'lov turi bo'yicha")
sns.swarmplot(data=sales, y="Payment Method", x="Total Revenue", hue="Product Category", size=5, ax=axes[2])
axes[2].set_title("To'lov turi va mahsulotning umumiy narxi mahsulot turi bo'yicha")
plt.show()
```

Eng arzon va narxi standart tovar kitoblar ekanligini bilib oldik

## Heatmap

```python
sales_cat_sold = pd.crosstab(sales["Product Category"], sales["Units Sold"], values=sales["Unit Price"], aggfunc="mean").round(0).fillna(0)
sales_cat_sold
```

```python
sales_cat_reg = pd.crosstab(sales["Product Category"], sales["Region"], values=sales["Unit Price"], aggfunc="mean").round(0).fillna(0)
sales_cat_reg
```

```python
sns.heatmap(sales_cat_sold, annot = True, fmt = ".0f", cmap = "YlGnBu", linewidths=5)
```

Ko'k va ko'kga yaqin bo'yalganlar eng qimmat tovarlardir.

```python
sns.heatmap(sales_cat_reg, annot = True, fmt = ".0f", cmap = "YlGnBu", cbar = False, linewidth =2)
```

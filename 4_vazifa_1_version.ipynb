# Bemor haqida ma'lumot
## Bemor ID
### PatientID: Har bir bemorga tayinlangan noyob identifikator (5034 dan 7425 gacha).

## Demografiya
### Yoshi: Bemorlarning yoshi 5 yoshdan 80 yoshgacha.
### Jins: Bemorlarning jinsi, bu erda 0 erkak va 1 ayolni bildiradi.
### Etnik kelib chiqishi: Bemorlarning etnik kelib chiqishi quyidagicha kodlangan:
#### 0: Kavkaz
#### 1: afro-amerikalik
#### 2: Osiyo
#### 3: Boshqa
### Ta'lim darajasi: Bemorning ta'lim darajasi, quyidagi kodlangan:
#### 0: Yo'q
#### 1: Oliy maktab
#### 2: Bakalavr darajasi
#### 3: Yuqori
## Turmush tarzi omillari
### BMI: bemorlarning tana massasi indeksi 15 dan 40 gacha.
### Chekish: Chekish holati, bu erda 0 "yo'q" va "1" "ha" degan ma'noni anglatadi.
### Jismoniy faollik: 0 dan 10 gacha bo'lgan soatlarda haftalik jismoniy faollik.
### DietQuality: dietaning sifati 0 dan 10 gacha.
### SleepQuality: 4 dan 10 gacha bo'lgan uyqu sifati reytingi.
## Atrof-muhit omillari va allergiya
### PollutionExposure: Ifloslanishga ta'sir qilish, 0 dan 10 gacha ball.
### Pollen ta'siri: 0 dan 10 gacha bo'lgan polen ta'siri.
### Changga ta'sir qilish: 0 dan 10 gacha baholangan changga ta'sir qilish.
### Uy hayvonlari allergiyasi: uy hayvonlari allergiyasi holati, bu erda 0 "yo'q" va "1" "ha" degan ma'noni anglatadi.
## Kasallik tarixi
### FamilyHistoryAstma: Oilaviy astma tarixi, bu erda 0 "yo'q" va "1" "ha" degan ma'noni anglatadi.
### HistoryOfAllergies : Allergiya tarixi, bu erda 0 yo'q va 1 "ha" degan ma'noni anglatadi.
### Egzema: ekzemaning mavjudligi, bu erda 0 yo'q va 1 "ha" degan ma'noni anglatadi.
### HayFever: pichan isitmasi bor, bu erda 0 yo'q va 1 ha degani.
### Gastroezofagial reflyuks: gastroezofagial reflyuksning mavjudligi, bu erda 0 "yo'q" va 1 "ha" degan ma'noni anglatadi.
## Klinik o'lchovlar
### LungFunctionFEV1: 1 soniyada majburiy ekspiratsiya hajmi (FEV1) 1,0 dan 4,0 litrgacha.
### LungFunctionFVC: majburiy hayotiy sig'im (FVC) 1,5 dan 6,0 litrgacha.
## Alomatlar
### Wheezing: xirillashning mavjudligi, bu erda 0 "yo'q" va 1 "ha" degan ma'noni anglatadi.
### ShortnessOfBreath: nafas qisilishining mavjudligi, bu erda 0 "yo'q" va 1 "ha" degan ma'noni anglatadi.
### Ko'krak qafasining siqilishi: ko'krak qafasidagi siqilishning mavjudligi, bu erda 0 "yo'q" va 1 "ha" degan ma'noni anglatadi.
### Yo'tal: yo'talning mavjudligi, bu erda 0 "yo'q" va 1 "ha" degan ma'noni anglatadi.
### Tungi simptomlar: tungi simptomlarning mavjudligi, bu erda 0 yo'q va 1 ha degani.
### Ishga tushirilgan mashqlar: Jismoniy mashqlar tufayli alomatlar paydo bo'ladi, bu erda 0 "Yo'q" va 1 "Ha" degan ma'noni anglatadi.
## Diagnostik ma'lumotlar
### Tashxis: Nafas tashxisi holati, bu erda 0 "yo'q" va 1 "ha" degan ma'noni anglatadi.
## Maxfiy ma'lumotlar
### DoctorInCharge : Ushbu ustunda barcha bemorlar uchun "Dr_Confid" qiymatiga ega bo'lgan davolovchi shifokor haqidagi maxfiy ma'lumotlar mavjud.
